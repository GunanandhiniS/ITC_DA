```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips.csv')
df.head(1)
data = df
```

```python
sns.barplot(x='day', y='total_bill', data=data, hue='sex' ,palette='colorblind6')
plt.title('Bar Plot of Average Total Bill Amounts by Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill Amount')
plt.show()
```
